## Rating

```python
#We are intersted in the number part of the column. split the column and appennd the first element
df['rating'] = df['rating'].apply(lambda x: x.split(' ')[0])
```

```python
#replace all occurence of '-' in rating with null values and convert the column to float data type
df['rating'] = df['rating'].replace('-', np.nan).astype(float)
```

## Dealing with Missing Values

```python
#check for columns with missing values
df.isnull().sum()
```

```python
#Fill rating column with the mean value.
df['rating'] = df['rating'].fillna(df['rating'].mean())
```
